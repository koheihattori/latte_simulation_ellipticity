```python
def calculateEllipticity(x,y,z,r_min,r_max):
    a=1.
    b=1.
    c=1.
    r=0.
    
    b_a=1.
    c_a=1.
    #r_abc = numpy.sqrt(x**2 + (y/b)**2 + (z/c)**2)
    #filter_r = numpy.where((r_min<=r_abc) & (r_abc<r_max))

    for i in range(100):
        # initial #
        b_a_old = b_a*1.0
        c_a_old = c_a*1.0
        
        r_abc = numpy.sqrt(x**2 + (y/b_a)**2 + (z/c_a)**2)
        filter_r = numpy.where((r_min<=r_abc) & (r_abc<r_max))
    
        Ixx =   numpy.sum(x[filter_r]*x[filter_r]/(r_abc[filter_r]**2))
        Ixy = - numpy.sum(x[filter_r]*y[filter_r]/(r_abc[filter_r]**2))
        Ixz = - numpy.sum(x[filter_r]*z[filter_r]/(r_abc[filter_r]**2))
        Iyy =   numpy.sum(y[filter_r]*y[filter_r]/(r_abc[filter_r]**2))
        Iyz = - numpy.sum(y[filter_r]*z[filter_r]/(r_abc[filter_r]**2))
        Izz =   numpy.sum(z[filter_r]*z[filter_r]/(r_abc[filter_r]**2))
        
        I = numpy.array([[Ixx, Ixy, Ixz],
                         [Ixy, Iyy, Iyz],
                         [Ixz, Iyz, Izz]
                         ])
            
        w,v = numpy.linalg.eig(I)

        a = numpy.sqrt(numpy.max(w))
        b = numpy.sqrt(numpy.median(w))
        c = numpy.sqrt(numpy.min(w))
        
        b_a = b/a
        c_a = c/a
        if (len(r_abc[filter_r]) > 1):
            median_r_abc = numpy.median(r_abc[filter_r])
            max_r_abc    = numpy.max(r_abc[filter_r])
        else:
            median_r_abc = None
            max_r_abc    = None
            break
        #print(i, median_r_abc)

        # check #
        if ( (numpy.fabs(b_a - b_a_old) < 0.001) & (numpy.fabs(c_a - c_a_old) < 0.001) ):
            print(i, median_r_abc)
            print('++++++++')
            break
    
    return a,b,c,median_r_abc, max_r_abc

x = particles_agama_DM_r200[:,0]
y = particles_agama_DM_r200[:,1]
z = particles_agama_DM_r200[:,2]
#mass_DM_sph = part_DM['dark']['mass'][filter_sph]
#print(len(mass_DM_sph))
r = numpy.sqrt(x*x + y*y + z*z)
list_r = []
list_a = []
list_b = []
list_c = []

_R0=1.
for i in range(0,20):
    r_min = float(i  )*5#(numpy.power(2., 0.5*float(i)  )-1.) #float(i)
    r_max = float(i+1)*5#(numpy.power(2., 0.5*float(i+1))-1.)
    print('###')
    print(i)
    print(r_min, r_max)
        
    a,b,c,median_r_abc, max_r_abc =calculateEllipticity(x,y,z,r_min,r_max)
    
    list_a.append(a)
    list_b.append(b)
    list_c.append(c)
    list_r.append(median_r_abc)
    print(a,b,c,median_r_abc, max_r_abc)
    print(c/a, b/a)

#print(list_r)
#print(list_a)
#print(len(list_r))
#print(len(list_a))
#print(len(list_b))
#print(len(list_c))




#pyplot.plot(numpy.array(list_r), numpy.array(list_b)/numpy.array(list_a), 'r-')
#pyplot.plot(numpy.array(list_r), numpy.array(list_b)/numpy.array(list_a), 'rx', label=r'$b/a$')

pyplot.plot(numpy.array(list_r), numpy.array(list_c)/numpy.array(list_b), 'r-')
pyplot.plot(numpy.array(list_r), numpy.array(list_c)/numpy.array(list_b), 'rx', label=r'$c/b$')
pyplot.plot(numpy.array(list_r), numpy.array(list_c)/numpy.array(list_a), 'b-')
pyplot.plot(numpy.array(list_r), numpy.array(list_c)/numpy.array(list_a), 'bo', label=r'$c/a$')
pyplot.legend(fontsize=20)
#pyplot.title(r'$N=4\times 10^5$ stars,  $N_\mathrm{orb}=5000$', fontsize=20)

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'ellipsoidal radius $/ (\mathrm{kpc})$', fontsize=20)
pyplot.ylabel(r'axis-ratio', fontsize=20)

pyplot.tight_layout()
pyplot.savefig('%s/m12m_ellipticity.png' % (dir_save))

```
